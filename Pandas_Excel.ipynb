```python
import pandas as pd
```

```python
#Lendo arquivos em excel
df1 = pd.read_excel("/content/drive/MyDrive/DIO.me/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/DIO.me/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/DIO.me/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/DIO.me/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/DIO.me/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Salvador.xlsx")
```

```python
#Juntando os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Alterando o tipo de dado de uma coluna
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
#Veriifcando linhas em branco
df.isnull().sum()
```

```python
#Criando coluna, no exemplo foi criado a coluna Receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Utilizando o nlargest para retornar, no exemplo o TOP 3 das receitas no arquivo.
df.nlargest(3,"Receita")
```

```python
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando um conjunto de dados.
df.sort_values("Receita", ascending=False).head(20)
```

```python
#Alterando a coluna Data para o tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
#Transformando a culuna Data para o formato data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Realizando agrupamento da Receita por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando as Colunas Ano_Venda , Mes_venda, Dia_Venda extraindo esses valores da coluna Data.
df["Ano_Venda"], df["Mes_Venda"], df["Dia_Venda"] = (df["Data"].dt.year, df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(10)
```

```python
#Filtrando vendas do ano de 2019 do mês de Março
vendas_marco_2019 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
